```python
import requests
import matplotlib.pyplot as plt

# Clave API de OpenWeatherMap
#api_key = 'API_KEY'

# Lista de ciudades de ejemplo
ciudades = ['Nueva York', 'México', 'Río de Janeiro', 'Mumbai', 'Dubái', 'Adís Abeba']

# Listas para almacenar datos de temperatura y humedad
temperaturas = []
humedades = []

# Obtener datos climáticos para las ciudades
for ciudad in ciudades:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={ciudad}&appid={api_key}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        temperatura = data["main"]["temp"]
        humedad = data["main"]["humidity"]
        temperaturas.append(temperatura)
        humedades.append(humedad)

# Encontrar el país más caliente y más húmedo
pais_mas_calor = ciudades[temperaturas.index(max(temperaturas))]
pais_mas_humedad = ciudades[humedades.index(max(humedades))]

# Datos para el gráfico de pastel
labels = ['País más caliente', 'País más húmedo']
values = [max(temperaturas), max(humedades)]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # Destacar la primera rebanada

# Crear el gráfico de pastel
plt.pie(values, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')

# Mostrar el gráfico
plt.show()

print(f'País más caliente: {pais_mas_calor}')
print(f'País más húmedo: {pais_mas_humedad}')
```
