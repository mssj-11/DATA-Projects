# **Análisis estadístico**

Generar datos aleatorios y realizar un análisis estadístico básico, incluyendo la creación de histogramas y gráficos de dispersión:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios para dos variables (como: edad y puntuación)
np.random.seed(0)
edad = np.random.randint(18, 60, 100)  # 100 edades aleatorias entre 18 y 60
puntuacion = np.random.normal(70, 10, 100)  # 100 puntuaciones con media 70 y desviación estándar 10

# Crear un histograma de edades
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(edad, bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Crear un gráfico de dispersión de edad vs. puntuación
plt.subplot(1, 2, 2)
plt.scatter(edad, puntuacion, color='salmon', alpha=0.7)
plt.title('Gráfico de Dispersión de Edad vs. Puntuación')
plt.xlabel('Edad')
plt.ylabel('Puntuación')

plt.tight_layout()
plt.show()
```

##### Este código generará un histograma de las edades y un gráfico de dispersión que muestra la relación entre la edad y la puntuación.

# **Pruebas de hipótesis, regresión y análisis de varianza**

### **Pruebas de hipótesis:**

Supongamos que deseas realizar una prueba de hipótesis para verificar si hay una diferencia significativa en las puntuaciones entre dos grupos de edades (menores de 30 y mayores de 30).
Utilizando la biblioteca scipy para realizar una prueba t de Student.

```python
import scipy.stats as stats

# Dividir los datos en dos grupos
edad_menores_30 = puntuacion[edad < 30]
edad_mayores_30 = puntuacion[edad >= 30]

# Realizar la prueba de t de Student para comparar los dos grupos
t_stat, p_value = stats.ttest_ind(edad_menores_30, edad_mayores_30)
if p_value < 0.05:
    print("Hay una diferencia significativa en las puntuaciones entre los grupos.")
else:
    print("No hay una diferencia significativa en las puntuaciones entre los grupos.")
```

## **Regresión:**

Supongamos que deseas realizar un análisis de regresión para modelar la relación entre la edad y la puntuación. Utilizando la biblioteca scipy para ajustar un modelo lineal.

```python
from scipy.stats import linregress

# Realizar un análisis de regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(edad, puntuacion)

# Imprimir los resultados del análisis de regresión
print("Coeficiente de pendiente:", slope)
print("Intercepto:", intercept)
print("Coeficiente de correlación (r):", r_value)
print("Valor p:", p_value)
```

## **Análisis de varianza (ANOVA):**
Supongamos que deseas realizar un análisis de varianza para determinar si existe una diferencia significativa en las puntuaciones entre tres o más grupos de edades. Utilizando la biblioteca scipy para realizar un ANOVA unidireccional.

```python
from scipy.stats import f_oneway

# Supongamos que tienes tres grupos de edades con datos válidos
edad_grupo_1 = [25, 28, 30, 33, 29]
edad_grupo_2 = [40, 42, 38, 37, 39]
edad_grupo_3 = [55, 58, 53, 50, 52]

# Realizar un análisis de varianza (ANOVA)
f_stat, p_value = f_oneway(edad_grupo_1, edad_grupo_2, edad_grupo_3)

# Imprimir el resultado del análisis de varianza
if p_value < 0.05:
    print("Hay una diferencia significativa en las puntuaciones entre los grupos de edades.")
else:
    print("No hay una diferencia significativa en las puntuaciones entre los grupos de edades.")
```
